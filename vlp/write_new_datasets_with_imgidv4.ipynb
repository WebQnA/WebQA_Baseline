```python
image_id_map_0328 = pickle.load(open(os.path.join(data_dir, "image_id_map_0328.pkl"), "rb"))
print(len(image_id_map_0328))
```

```python
dataset = json.load(open("/home/yingshac/CYS/WebQnA/WebQnA_data_new/WebQA_0904_concat_newimgid_newguid.json", "r"))
print(Counter([dataset[k]['split'] for k in dataset]))
print(len(set([dataset[k]['Guid'] for k in dataset])))
print(Counter([dataset[k]['Qcate'] for k in dataset]))
```

```python
for g in dataset:
    d = dataset[g]
    if 'img_posFacts' in d:
        for im in d['img_posFacts']:
            imgidv3 = im['image_id']
            im['image_id'] = image_id_map_0328[imgidv3]
    for im in d['img_negFacts']:
        imgidv3 = im['image_id']
        im['image_id'] = image_id_map_0328[imgidv3]

```
